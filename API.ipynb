```python
import requests
```

```python
# 1. Obtain a token
token_response = requests.post("http://127.0.0.1:5000/auth/token", json={
    "username": "admin",
    "password": "admin123"
})
token = token_response.json()["token"]
print("Token obtained :", token[:20] + "...")
```

```python
# 2. Call the API using the token
headers = {"Authorization": f"Bearer {token}"}

r = requests.get("http://127.0.0.1:5000/api/sephora/products", headers=headers, params={"limit": 3})
print("\nSephora products:", r.json())

r = requests.get("http://127.0.0.1:5000/api/sephora/brands", headers=headers)
print("\nTop brands:", r.json())

r = requests.get("http://127.0.0.1:5000/api/comparaison", headers=headers)
print("\nComparison:", r.json())
```
