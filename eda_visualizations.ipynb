```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/processed/sephora_clean.csv')
```

```python
# 1. Price Distribution of Sephora Products
plt.figure(figsize=(10, 6))
plt.hist(df['price_usd'], bins=50, edgecolor='black')
plt.title('Price Distribution of Sephora Products')
plt.xlabel('Price (USD)')
plt.ylabel('Number of products')
plt.savefig('Visualizations/price_distribution.png')
plt.show()
```

```python
# 2. Top 10 brands by number of products
top_brands = df['brand_name'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_brands.plot(kind='barh')
plt.title('Top 10 brands by number of products')
plt.xlabel('Number of Products')
plt.tight_layout()
plt.savefig('Visualizations/top_brands.png')
plt.show()
```

```python
df2 = pd.read_csv('data/processed/sephora_dashboard_data.csv')
```

```python
# 3. Pourcentage de produits avec ingrdients restreints par type
type_stats = df2.groupby('product_type').agg({
    'has_restricted_ingredient': 'mean'
}).sort_values('has_restricted_ingredient', ascending=False) * 100

plt.figure(figsize=(10, 6))
type_stats.plot(kind='bar', color='coral')
plt.title('% of products with restricted ingredients')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Visualizations/restricted_by_type.png')
plt.show()
```

```python
# 4. Price vs rating correlation
plt.figure(figsize=(8, 6))
plt.scatter(df['price_usd'], df['rating'], alpha=0.5)
plt.title('Price vs Rating Correlation')
plt.xlabel('Prix (USD)')
plt.ylabel('Rating')
plt.savefig('Visualizations/price_vs_rating.png')
plt.show()
```
